INTRODUCTION TO  NUMPY

NumPy is one of the most important foundational packages for numerical computing in Python

It provides a high performance multidimensional array object and tools for working with these arrays.

**DIFFERENCE B/W ITEMSIZE AND SIZE**

The itemsize attribute in NumPy provides information about the memory consumption of individual elements within a NumPy array.This value is determined by the data type (dtype) of the array. 

The size attribute in NumPy returns the total number of elements in a NumPy array. This attribute provides a single integer value representing the count of all elements within the array, regardless of its dimensions or shape. 



```python
import numpy 
Sneha=numpy.array([[21,23,22,21,24],[56,36,78,88,78]],dtype=float)
#my array=numpy.array([1,2,3,4])
print(Sneha)
print(Sneha.shape)#row,col
print(Sneha.size) # no of elements
print(Sneha.itemsize)#memory used by one element(datatype of array)
print(Sneha.ndim)#dimensions 1d,2d ,3d
```

```python
import numpy as np
print(np.full((2,2),7))
print(np.eye(3))#identity matrix of order 3
print(np.arange(0,10,2))#starting value,ending value(excluded),stepsize
print(np.linspace(0,2,5))#starting value,ending value(included),no of values having same difference
print(np.random.rand(2,3))#taking float values b/w 0 and 1
print(np.random.randint(0,10,(3,3)))#returns integer values between 0 and 10 
print(np.random.seed(23))
print(np.random.randint(0,10,(3,3)))

```

*AXIS HANDLING*****

Axis 0:operates down columns

Axis 1:operates across rows

```python
import numpy as np
arr1=np.array([[1,2,3],[3,4,5]])
print('column wise sum:',arr1.sum(axis=0))
print('crow wise mean:',arr1.mean(axis=1))
print('max of each row:',arr1.max(axis=1))
print('Transpose:\n',arr1.T)
print('Reshape to 1D:',arr1.flatten())#converts every 2d or 3d matrix/array into 1D array 
```
